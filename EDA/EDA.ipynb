# Dropping Rows with Missing Data

A lot of the subjects have missing data, indicated often by a value of 999 (see the abornmal spikes in the histograms), so before we start to look at the attributes of the data, it may be helpful to clean the missing values to get a better sense of the underlying relationships between our parameters.

```python
data = mela[attribs]
new_df = data[(data != 999).all(1)]
#new_df = new_df[new_df['RACE'] != 99]
#new_df = new_df[new_df['SPANISH_HISPANIC_ORIGIN'] != 9]
new_df = new_df[new_df['MITOSES'] != 988]
new_df = new_df[new_df['DEPTH'] != 0]
new_df = new_df[new_df['DEPTH']<= 980]
new_df = new_df.dropna()


new_df.info()
new_df.head()
```

# Combining Feature Values

In addition, there were several coding schemes used for the data, so for some of our attributes, it is useful to combine these into a single category. In addition, some of our features get very granular, where we only have a single representative in the dataset with those demographics (for example Hmong in `RACE`). For now, I consider a more coarse-grained separation of these features.

```python
import numpy as np

new_df['ULCERATION'] = new_df['ULCERATION'].replace(10.0, 1.0)

new_df['CS_EXTENSION'].value_counts()
new_df['CS_EXTENSION'] = new_df['CS_EXTENSION'].replace([
    310, 315, 320, 330, 335, 340, 350, 355, 360, 370, 375, 380, 400, 800, 950],
    [999, 999, 999, 999, 999, 999, 999, 999, 999, 999, 999, 999, 999, 999, 999])
new_df = new_df[(new_df != 999).all(1)]
new_df['CS_EXTENSION'].value_counts()

new_df['LYMPH_NODE_METS'] = new_df['LYMPH_NODE_METS'].replace([5.0, 10.0], [0, 1])
new_df = new_df[new_df['LYMPH_NODE_METS']<= 1.5]
new_df['LYMPH_NODE_METS'].value_counts()

#new_df['TNM_PATH_N'].value_counts()
#new_df = new_df[new_df['TNM_PATH_N'] != '88']
#new_df = new_df[new_df['TNM_PATH_N'] != 'pX']
#new_df['TNM_PATH_N'].value_counts()
#new_df['TNM_PATH_N'] = new_df['TNM_PATH_N'].apply(lambda x: 0 if x == 'p0' else 1)
#new_df['TNM_PATH_N'] = new_df['TNM_PATH_N'].apply(lambda x: x if x == 0 or x == 1 else 1)

#new_df['RACE'].value_counts()
#a = np.array(new_df['RACE'].values.tolist())
#new_df['RACE'] = np.where(a > 2, 3, a).tolist()
#new_df['RACE'].value_counts()#new_df['RACE'] = new_df['RACE'].apply(lambda x: [y if y <= 1 else 2 for y in x])
#new_df['RACE']

#a = np.array(new_df['SPANISH_HISPANIC_ORIGIN'].values.tolist())
#new_df['SPANISH_HISPANIC_ORIGIN'] = np.where(a > 1, 1, a).tolist()
#new_df['SPANISH_HISPANIC_ORIGIN'].value_counts()

#new_df.describe()
```

```python
#plt.rcParams.update({'font.size': 30})
#plt.title('Metastasis')
def sephist(col):
    yes = new_df[new_df['LYMPH_NODE_METS'] == 1.0][col]
    no = new_df[new_df['LYMPH_NODE_METS'] == 0.0][col]
    return yes, no

for num, alpha in enumerate(['AGE', 'DEPTH']):
    plt.subplot(2, 1, num+1)
    plt.hist(sephist(alpha)[0], bins=25, alpha=0.5, label='POSITIVE', color='b')
    plt.hist(sephist(alpha)[1], bins=25, alpha=0.5, label='NEGATIVE', color='r')
    plt.legend(loc='upper right')
    plt.title(alpha)
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)

plt.rcParams.update({'font.size': 30})
new_df['DEPTH'].hist(bins=20,by=new_df['LYMPH_NODE_METS'],figsize=(20,15))
plt.rcParams.update({'font.size': 5})
```
